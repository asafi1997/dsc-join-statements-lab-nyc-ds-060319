## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import pandas as pd
import sqlite3
conn = sqlite3.connect('data.sqlite')
c = conn.cursor()
```

## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
#Your code here
c.execute("""SELECT offices.city, employees.firstName, employees.lastName
                FROM employees
                JOIN offices
                USING (officeCode)
                WHERE city = 'Boston';""")
d = pd.DataFrame(c.fetchall())
d.columns = [x[0] for x in c.description]
d
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
#Your code here
c.execute("""SELECT city, COUNT(*)
                FROM offices
                LEFT JOIN employees
                USING (officeCode)
                GROUP BY city;""")
d = pd.DataFrame(c.fetchall())
d.columns = [x[0] for x in c.description]
d
```

## Level Up: Display the names of each product each employee has sold

```python
# Your code here
c.execute("""SELECT products.productName, employees.firstName, employees.lastName
                FROM employees
                JOIN customers
                ON (employees.employeeNumber = customers.salesRepEmployeeNumber)
                JOIN orders 
                USING (customerNumber)
                JOIN orderdetails
                USING (orderNumber)
                JOIN products
                USING (productCode)""")
d = pd.DataFrame(c.fetchall())
d.columns = [x[0] for x in c.description]
d
```

## Level Up: Display the Number of Products each employee has sold

```python
#Your code here
c.execute("""SELECT  employees.firstName, employees.lastName,
                COUNT(employees.employeeNumber) as NumPRo
                FROM employees
                JOIN customers
                ON (employees.employeeNumber = customers.salesRepEmployeeNumber)
                JOIN orders 
                USING (customerNumber)
                JOIN orderdetails
                USING (orderNumber)
                JOIN products
                USING (productCode)
                GROUP BY employeeNumber; """)
d = pd.DataFrame(c.fetchall())
d.columns = [x[0] for x in c.description]
d
```
